```python
def calculate_pity (df: pd.DataFrame(), col_name: str):
    pities = []
    count_4s = 0
    count_5s = 0

    for row in df.values:
        
        if row[3] == 3:
            pity = 1
            count_4s = count_4s + 1
            count_5s = count_5s + 1

        elif row[3] == 4:
            pity = count_4s + 1
            count_4s = 0
            count_5s = count_5s + 1

        elif row[3] == 5:
            pity = count_5s + 1
            count_4s = count_4s + 1
            count_5s = 0
        else:
            pity = 99
            count_4s = 0
            count_5s = 0
        
#         for debugging purposes only
#         print(row, pity, count_4s, count_5s)

        pities.append(pity)

    df[col_name] = pities
    
    return df
```

```python
players = [('Алина', 'Alina крутки'), 
          ('Алена', 'Alena wishes'),
          ('Никита', 'История круток Никита'),
          ('Гия', 'Гия крутки')]

sheets = ['Character Event', 'Standard']

df = pd.DataFrame()

for player in players:
    player_name = player[0]
    file_name = player[1]
#     print(player_name)
    
    for sheet in sheets:
        print(sheet)
        temp_df = pd.read_excel(f'/Users/alinacepurnova/Downloads/{file_name}.xlsx', 
                              sheet_name = sheet)
        temp_df['Player'] = player_name
        temp_df['Event'] = sheet
        temp_df = temp_df.sort_values(['Event', 'Player', 'Time'])
        
        temp_df = calculate_pity(temp_df, 'Calculated Pity')
        
        df = pd.concat([temp_df, df])

df = df.sort_values(['Event', 'Player', 'Time'])
print(len(df))
df.head(3)
```
